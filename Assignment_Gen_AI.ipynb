# Title: Build an Instruction-Following Chatbot using
# google/flan-t5-base with Custom Instructions

```python
model_id='google/flan-t5-base'
```

# Required Libraries:-

```python
!pip install transformers
```

```python
from transformers import pipeline
```

# Model Building:-

```python
chatbot=pipeline('text2text-generation',model=model_id)
```

# Defining Communication:-

```python
def chat(user_input,user_role):
  prompt=f'As a {user_role}, please answer the following:{user_input}'
  reply=chatbot(prompt,max_new_tokens=100)[0]['generated_text']
  return reply
```

```python
print('Hii Surya...its your BOT..let"s \n if you dont want press exit..')
while True:
  inp=input('You:')
  if inp.lower() in 'exit':
    print('OK....byy..\nThank You')
    break
  else:
    role=input("Enter your role:-")
    reply=chat(inp,role)
    print('Bot:',reply)
```
