CREATING A CHAT BOT WITH 'google/flan-t5-small'

# INPORTING PIPELINE:

```python
from transformers import pipeline
```

# MODEL CREATION:

```python
chatbot=pipeline('text2text-generation',model='google/flan-t5-small')
```

# DEFINING THE FUNCTIONS FOR INTERECTION:-

```python
def responce(prompt):
  result=chatbot(f"Answer the following prompt:\n\n{prompt}",max_new_tokens=100)
  return result[0]['generated_text']
```

```python
while True:
  prompt=input("You:")
  if prompt.lower() in ['exit','leave','end','stop']:
    print('By By....\nRemember me anythime.')
    break
  ans=responce(prompt)
  print('Bot:',ans)
```

# We can see the model is very much besic,not able to give proper answer.
