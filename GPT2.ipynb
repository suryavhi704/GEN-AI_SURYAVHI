```python
# importing  pipeline
from transformers import pipeline
```

```python
# Model Creation
model=pipeline('text-generation',model='gpt2')
```

```python
prompt='What is ai?' #selected prompt
```

```python
prediction = model(prompt,max_length=50,num_return_sequences=1) #model computation
```

```python
prediction[0] #model generated text
```

```python
prompt="What is Quantum computing?" #selected prompt
```

In artificial intelligence (AI) and machine learning, temperature is a parameter for adjusting the output of large language models (LLMs). Temperature controls the randomness of text that is generated by LLMs during inference.

```python
prediction = model(prompt,max_length=50,num_return_sequences=1,temperature=0.8)
```

```python
prediction[0]
```

```python
# TOKENIZER OF GPT
from transformers import GPT2Tokenizer
```

```python
token=GPT2Tokenizer.from_pretrained('gpt2') #tokenizer object
```

```python
sen='hi,i am very happy'
```

```python
# encoding:
ans=token.encode(sen)
```
