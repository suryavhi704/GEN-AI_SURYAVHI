# Mistral-instruct-7B  --> GPU + Private model

Importing libraries:-

```python
from transformers import AutoTokenizer,AutoModelForCausalLM,pipeline
```

Model information:-

```python
# mistralai/Mistral-7B-Instruct-v0.1
# my token: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
model_id='mistralai/Mistral-7B-Instruct-v0.1'
```

Getting the model token:

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_id)
```

```python
model=AutoModelForCausalLM.from_pretrained(model_id)
```

Pipelining the model with type and tokenizer:-

```python
chatbot=pipeline('text-generation',tokenizer=tokenizer,model=model)
```

```python
def chat():
  print('Hey...im your chatbot.enter "Exit" it you dont want to continue:-')
  while True:
    prompt=input('You:')
    if prompt.lower() in 'exit':
      print('By..By..Have a good day...')
      break
    prompt=f"[INST]{prompt}[/INST]"
    ans=chatbot(prompt,max_new_tokens=100,do_sample=True)[0]['generated_text']
    responce=ans.split("[/INST]")[-1].strip()
    print('Bot:',responce)
```

```python
chat()
```

# Google Falcon Chat Bot:-

Model id:-

```python
model_id="tiiuae/falcon-rw-1b"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
model=AutoModelForCausalLM.from_pretrained(model_id)
```

```python
chatbot=pipeline('text-generation',model=model,tokenizer=tokenizer)
```

```python
def chat(user_input):
  prompt=f"<s>User: {user_input}\nAI:"
  responce=chatbot(
      prompt,
      max_new_tokens=200,
      do_sample=True,
      temperature=0.7,
      pad_token_id=tokenizer.eos_token_id
  )
  return responce[0]['generated_text']
```

```python
while True:
  print('Hii..its falcon this side....\nlets chat\nIf you dont want then enter "Exit"')
  inp=input("You:")
  if inp.lower() in 'exit':
    print('By..By..see you soon.')
    break
  response=chat(inp)
  print("Bot:",response)
```
